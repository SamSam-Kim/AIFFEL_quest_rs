# Modeling

- 각 모델에 보다 적합한 파라미터 설정을 위한 Grid Search 진행

```python
def best_GridSearch(model, train, y, param_grid):
    grid_model = GridSearchCV(model, param_grid=param_grid, \
                        scoring='neg_mean_squared_error', \
                        cv=5, verbose=1, n_jobs=10)
    
    grid_model.fit(train, y)
    
    params = grid_model.cv_results_['params']
    score = grid_model.cv_results_['mean_test_score']
    
    results = pd.DataFrame(params)
    results["score"] = score
    results['RMSLE'] = np.sqrt(-1 * results['score'])
    results.sort_values(by="RMSLE", ignore_index=True, inplace=True)
    
    return results
```

```python
param_grid = {
    'n_estimators': [100, 150],
    'max_depth': [5, 7, 10],
}
```

```python
model = LGBMRegressor(random_state=random_state)
LGBM_result = best_GridSearch(model, X, y, param_grid)
LGBM_result
```

```python
model = GradientBoostingRegressor(random_state=random_state)
GB_result = best_GridSearch(model, X, y, param_grid)
GB_result
```

```python
model = XGBRegressor(random_state=random_state)
XGB_result = best_GridSearch(model, X, y, param_grid)
XGB_result
```

```python
model = RandomForestRegressor(random_state=random_state)
RF_result = best_GridSearch(model, X, y, param_grid)
RF_result
```

- Random Forest의 성능이 비교적 떨어져서 제외하고 나머지 모델을 Grid Search로 찾은 파라미터를 적용해서 초기화

```python
lgbm = LGBMRegressor(max_depth=int(LGBM_result.loc[0][0]),
                     n_estimators=int(LGBM_result.loc[0][1]),
                     random_state=random_state)

gboost = GradientBoostingRegressor(max_depth=int(GB_result.loc[0][0]),
                     n_estimators=int(GB_result.loc[0][1]),
                     random_state=random_state)

xgboost = XGBRegressor(max_depth=int(XGB_result.loc[0][0]),
                     n_estimators=int(XGB_result.loc[0][1]),
                     random_state=random_state)

models = [gboost, xgboost, lgbm]
```

```python
def get_scores(models, X_train, X_test, y_train, y_test):
    df = {}

    for model in models:
        model.fit(X_train, y_train)
        predictions = model.predict(X_test)

        model_name = model.__class__.__name__
        df[model_name] = np.sqrt(mean_squared_error(np.expm1(y_test), np.expm1(predictions)))

    score_df = pd.DataFrame(df, index=['RMSE']).T.sort_values('RMSE', ascending=False)

    return score_df
```

- 다시 한 번 너무 튀는 점수가 없는지 확인

```python
get_scores(models, X_train, X_test, y_train, y_test)
```

```python
def AveragingBlending(models, X, y, sub):
    for m in models : 
        m.fit(X, y)
       
    predictions = np.column_stack([
        np.expm1(m.predict(sub)) for m in models
    ])
    
    return np.mean(predictions, axis=1)
```

```python
final_pred = AveragingBlending(models, X, y, sub)
```

# Feature Engineering

## New Features

**Coverage: 유사 건폐율**

sqft_living & sqft_lot : 주거 공간과 부지의 평방 피트

- 해당 데이터를 활용하여 건폐율 값을 유사하게 계산 가능
    - 건폐율: 토지 면적 대비 건축물 면적
    - 지역에 따라 다르지만 일반적으로 건폐율이 낮으면 집 값이 높아지는 경향이 있다고 함.

- 즉, 건폐율이 집 값에 유의미한 영향을 준다고 할 수 있으므로 건폐율 컬럼을 추가.

```python
data
```

```python
data["coverage"] = data.sqft_living / data.sqft_lot
data.tail()
```

**Coverage15: 근처 15 가구의 유사 건폐율**

위와 같은 맥락으로 sqft_living15 & sqft_lot15 데이터를 활용해 컬럼 추가

```python
data["coverage15"] = data.sqft_living15 / data.sqft_lot15
data.head()
```

**avg_ppf 피트 당 가격(이하 핏당가)의 평균**

- 지역에 따라 집 값에 영향이 있으므로 지역을 표현하는 값이 필요함.

- zipcode로 지역을 구분할 수 있으나 zipcode의 숫자는 수치를 의미하지 않아 입력값으로 적절하지 않음.

- 부동산에서 지역에 따라 '평당가'를 말하는 것에서 착안하여 가지고 있는 zipcode 정보와 price 정보를 활용하여 zipcode로 구분되는 지역별 평균 핏당가 컬럼을 추가.

```python
len(data["zipcode"].unique()), len(train["zipcode"].unique()), len(sub_data["zipcode"].unique())
```

- train 데이터와 sub 데이터는 같은 zipcode를 공유하고 있으므로 price 정보를 알고있는 train 데이터에서 평균 핏당가를 계산하여 sub 데이터에도 추가 가능.

```python
train
```

```python
# Price 정보가 포함된 train 데이터만으로 zipcode별 평균 집값 구하기

avg_ppf = {} # 평균 집값 정보를 저장할 딕셔너리 초기화

for i in train["zipcode"].unique():
    # zipcode 별 핏당가. [price / sqft_living]
    ppf = train.loc[train["zipcode"]==i, 'price'] / train.loc[train["zipcode"]==i, 'sqft_living']
    
    # zipcode 별 평균 핏당가 계산 후 저장
    z_avg_ppf = ppf.mean()
    avg_ppf[i] = z_avg_ppf
    
avg_ppf
```

- zipcode별 평균 핏당가를 data에 적용

```python
# avg_ppf 컬럼 초기화
data["avg_ppf"] = 0
data
```

```python
# zipcode마다 해당하는 평균 핏당가를 입력
for k, v in avg_ppf.items():
    data.loc[train["zipcode"]==k, 'avg_ppf'] = v

# 입력 결과 확인
data
```

# 회고

**Feature Engineering은 정말 어렵다...**
- 위에 기록된 내용 말고도 정말 다양한 시도를 했다.
    - 시간이 너무 지체되면서 서버와의 연결이 끊겨 하던 걸 날린 것도 여러번.
    - 그러나 중요한 모든 시도에서 결과적으로 점수가 상승하지는 못했다는 것.
- 일반적인 생각으로 건폐율이나 지역별 평당가 등의 특성이 긍정적인 영향을 미칠 것이라고 생각했으나 전혀 그렇지 못했다.
    - 아마도 지역적 특색이 우리나라가 아닌 미국의 데이터이기 때문에 더 그런 것 같기도 하다.
    - 어쩌면 도메인 지식에 대해 어설프게 알고 확신하고 수정한 것들이 오히려 모델을 망가트릴 수도 있다는 점을 배운 것 같다.
    
    
**Modeling에 대한 이해가 더 필요하다.**
- 배운 내용을 활용해서 좀 다르게 적용해보려는 시도들에서 어지러울 정도로 헷갈렸다.
    - 근본적인 이해가 부족한 상태로 값만 바꾼다고 해결이 되지 않는 것 같다.
- 각각의 방식이 어떤 식으로 작용하고 각 파라미터들이 어떤 역할을 하는지 좀 더 이해한다면 보다 신속하고 쉽게 모델링 작업을 할 수 있었을 것 같다.
